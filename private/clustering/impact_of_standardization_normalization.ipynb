```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
X = np.random.standard_normal((50,2))
X[:25,0] += 30
X[:25,1] -= 4

print(f'std X1: {np.std(X[:,0])}')
print(f'std X2: {np.std(X[:,1])}')
print(f'mean X1: {np.mean(X[:,0])}')
print(f'mean X2: {np.mean(X[:,1])}')

plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=2, n_init=20).fit(X)
print(f'labels: {kmeans.labels_}')

plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.show()
```

Normalize

```python
from sklearn.preprocessing import StandardScaler

norm_scaler = StandardScaler(with_std=False, with_mean=True)
X_norm = norm_scaler.fit_transform(X)

print(f'std X1: {np.std(X_norm[:,0])}')
print(f'std X2: {np.std(X_norm[:,1])}')
print(f'mean X1: {np.mean(X_norm[:,0])}')
print(f'mean X2: {np.mean(X_norm[:,1])}')

plt.scatter(X_norm[:,0], X_norm[:,1])
plt.show()
```

```python
kmeans_norm = KMeans(n_clusters=4, random_state=2, n_init=20).fit(X_norm)
print(f'labels: {kmeans_norm.labels_}')

print(f'equal to unprocessed: {np.all(kmeans_norm.labels_ == kmeans.labels_)}')
plt.scatter(X_norm[:,0], X_norm[:,1], c=kmeans_norm.labels_)
plt.show()
```

Standardized

```python

std_scaler = StandardScaler(with_std=True, with_mean=False)
X_std = std_scaler.fit_transform(X)

print(f'std X1: {np.std(X_std[:,0])}')
print(f'std X2: {np.std(X_std[:,1])}')
print(f'mean X1: {np.mean(X_std[:,0])}')
print(f'mean X2: {np.mean(X_std[:,1])}')

plt.scatter(X_std[:,0], X_std[:,1])
plt.show()
```

```python
kmeans_std = KMeans(n_clusters=4, random_state=2, n_init=20).fit(X_std)
print(f'labels: {kmeans_std.labels_}')

print(f'equal to unprocessed: {np.all(kmeans_std.labels_ == kmeans.labels_)}')
plt.scatter(X_std[:,0], X_std[:,1], c=kmeans_std.labels_)
plt.show()
```

--> Kmeans is sensitive to standardization
